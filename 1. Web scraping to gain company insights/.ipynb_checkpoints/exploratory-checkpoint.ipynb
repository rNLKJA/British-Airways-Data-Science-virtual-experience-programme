```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import string
%matplotlib inline
```

```python
from pprint import pprint
```

```python
review_ratings_df = pd.read_csv("../data/review_ratings_df.csv")
review_ratings_df.tail()
```

```python
review_ratings_df['initial'] = [i.upper() for i in review_ratings_df.airline.str[0]]
review_ratings_df
```

```python
plt.bar(x = list(string.ascii_uppercase), height=review_ratings_df.initial.value_counts())
plt.title("No. Airlines (Based on initial)")
```

```python
plt.bar(x = list(string.ascii_uppercase), height=review_ratings_df[["initial", "reviews"]].groupby("initial").sum().reviews)
plt.title("No. reviews based on initial")
plt.show()
```

```python
df = review_ratings_df.sort_values(by="reviews", ascending=False).head(20)
df
```

```python
plt.barh(y=df.airline, width=df.reviews)
plt.title("Top 20 Airlines of No. Reviews")
```

```python
df = review_ratings_df[["score (max 10)"]].value_counts()
df
```

```python
plt.bar(x=np.arange(1,11,1), height=df.sort_index().values)
plt.title("No. Airline Rating")
```

```python
df = review_ratings_df.sort_values(by="score (max 10)", ascending=False).head(20)
df
```
